```python
 #funcion para convertir porcentaje a absoluto
def make_autopct(values): 
    def my_autopct(pct): 
     total = sum(values) 
     val = int(round(pct*total/100.0)) 
     return ' {v:d}'.format(p=pct,v=val) 
    return my_autopct
```

```python
plt.pie(values, labels={1:'potable',0:'no potable'},
        colors=['orange', 'steelblue'],
          textprops={'size': 'x-large'},
          autopct= make_autopct(values),
         wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'})
```

# regresion logistica

```python
dt_features=w_colombia.drop('Potability', axis=1)
dt_features
```

```python
dt_target=w_colombia['Potability']
dt_target
```

```python
dt_features = StandardScaler().fit_transform(dt_features)
X_train,X_test, y_train, y_test= train_test_split(dt_features,dt_target, test_size=0.35, random_state=100)
```

```python
pca = PCA(n_components=3)
pca.fit(X_train)
ipca = IncrementalPCA(n_components=9, batch_size=10)
ipca.fit(X_train)

```

```python
r=dt_features[90]
r
```

```python
%matplotlib inline
plt.plot(range(len(pca.explained_variance_)), pca.explained_variance_ratio_)
plt.show()
```

```python
logistic = LogisticRegression(solver='lbfgs')
dt_train= pca.transform(X_train)
dt_test= pca.transform(X_test)
logistic.fit(dt_train, y_train)
print(" Score PCA: ", logistic.score (dt_test, y_test))
```

```python
logistic = LogisticRegression(solver='lbfgs')
dt_train= ipca.transform(X_train)
dt_test=ipca.transform(X_test)
logistic.fit(dt_train, y_train)
print(' Score IPCA: ', logistic.score (dt_test, y_test))
```
